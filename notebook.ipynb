Die Schlüssel bestehen aus den drei Zahlen e, d und N  
  
N: RSA-Modul  
e: Verschlüsselungsexponenten  
d: Entschlüsselungsexponenten
  
öffentlicher Schlüssel: (e, N)  
privater Schlüssel:     (d, N)

```python
# Zwei Primzahlen

import random
import math
from sympy import isprime

# TODO: Ist das stochastisch unabhängig?

random.seed(0) # set seed for reproducability
while True:
    range = (1,100)
    p = random.randint(range[0], range[1])
    q = random.randint(range[0], range[1])
    criteria = abs(math.log2(p) - math.log2(q))

    if not isprime(p) or not isprime(q):
        continue

    if 0.1 < criteria < 30:
        break


print("criteria: ", criteria)
print("p: ", p)
print("q: ", q)
```

```python
N = p * q
print("N:", N)
```

```python
phi = (p - 1) * (q - 1)
print(phi)
```

```python
while True:
    e = random.randint(2, phi - 1)
    if math.gcd(e, phi) == 1:
        break
print("e: ", e)
```

```python
d = pow(e, -1, phi)
print("d: ", d)

result = (e * d) % 10 # das muss 1 ergeben
print("result: ", result)
```

```python
# Zusammenfassung der resultate

print("e: ", e)
print("d: ", d)
print("N: ", N)
```

```python
# Es soll die Zahl 7 verschlüsselt werden

klartext = 7
chiffrat = pow(klartext, e, N)
print("klartext: ", klartext)
print("chiffrat: ", chiffrat)
```

```python
# Entschlüsseln

reconstruct = pow(chiffrat, d, N)
print("reconstruct: ", reconstruct)
assert reconstruct == klartext, "Rekonstruktion entspricht nicht dem Klartext"
```
